```python
import numpy as np
import pandas as pd
```

```python
housing = pd.read_csv('housing.csv')
```

```python
housing.head()
```

1. Масштабирование:

а. Нормализовать признак - median_income

```python
from sklearn import preprocessing
```

```python
# normalize the "median_income" attribute
norm = preprocessing.MinMaxScaler()
housing["median_income_normalized"] = norm.fit_transform(housing[["median_income"]])
```

```python
housing["median_income_normalized"]
```

б. Привести признак - population к масштабу от 0 до 100


```python

# set 0-100 the "population" attribute
to_0_100 = preprocessing.MinMaxScaler(feature_range=[0,100])
housing["population_to_0_100"] = to_0_100.fit_transform(housing[["population"]])
```

```python
housing["population_to_0_100"]
```

2. Заполнение пропусков:

а. Заполнить пропуски в столбце total_bedrooms средними значениями по категориям столбца ocean_proximity

Например, если пропущенно значение в строке с категорией INLAND, нужно заполнить его средним значением по этой категории

```python
# replacement of missing values in column "total_bedrooms" on the avg for each category of column "ocean_proximity"
housing['total_bedrooms']= housing.groupby("ocean_proximity")['total_bedrooms'].transform(lambda x: x.fillna(x.mean()))
housing.info()
```

3. Категорийные атрибуты:

а. Посчитать среднее кол-во комнат в строке (total_rooms/households)

```python
# count avg qty rooms per house
housing['avg_rooms_per_house'] = (housing['total_rooms']/housing['households'])
housing.head()
```

б. Разбить полученный столбец на группы на ваше усмотрение (Например, 1 комната, 2 комнаты, 3 и более комнат)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
housing["avg_rooms_bin"] = housing["avg_rooms_per_house"]\
.apply(lambda x: '<3' if x < 3 else '3-6' if x < 7 else '>6')
```
